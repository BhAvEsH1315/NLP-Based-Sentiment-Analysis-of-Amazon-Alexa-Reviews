# Preprocessing and Modelling

To build the corpus from the 'verified_reviews' we perform the following - <br>
1. Replace any non alphabet characters with a space
2. Covert to lower case and split into words
3. Iterate over the individual words and if it is not a stopword then add the stemmed form of the word to the corpus

```python
corpus = []
stemmer = PorterStemmer()
for i in range(0, data.shape[0]):
  review = re.sub('[^a-zA-Z]', ' ', data.iloc[i]['verified_reviews'])
  review = review.lower().split()
  review = [stemmer.stem(word) for word in review if not word in STOPWORDS]
  review = ' '.join(review)
  corpus.append(review)
```

### Why Use the Porter Stemmer?

Search Optimization: Improves search performance by matching different forms of a word. For instance, a search query for "run" will also match "running" and "runner."

Text Analysis: Helps in analyzing and grouping similar words together, making text data more manageable.

Natural Language Processing: Reduces the dimensionality of text data, which can improve the performance of machine learning models.

### Why Use the CountVectorizer?

Text to Numerical Conversion: Converts textual data into a numerical format that can be used by machine learning models.

Feature Extraction: Extracts features (words) from text data, allowing the model to learn from the textual content.

Simple and Efficient: Easy to use and works efficiently with large datasets.

```python
cv = CountVectorizer(max_features = 2500)

#Storing independent and dependent variables in X and y
X = cv.fit_transform(corpus).toarray()
y = data['feedback'].values
```

```python
#Saving the Count Vectorizer
pickle.dump(cv, open('Models/countVectorizer.pkl', 'wb'))
```

Checking the shape of X and y

```python
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

Splitting data into train and test set with 30% data with testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 15)

print(f"X train: {X_train.shape}")
print(f"y train: {y_train.shape}")
print(f"X test: {X_test.shape}")
print(f"y test: {y_test.shape}")
```

```python
print(f"X train max value: {X_train.max()}")
print(f"X test max value: {X_test.max()}")
```

#### We'll scale X_train and X_test so that all values are between 0 and 1.

```python
scaler = MinMaxScaler()

X_train_scl = scaler.fit_transform(X_train)
X_test_scl = scaler.transform(X_test)
```

```python
#Saving the scaler model
pickle.dump(scaler, open('Models/scaler.pkl', 'wb'))
```

#### Random Forest
Random Forest is a powerful and reliable algorithm for both classification and regression tasks. Its ability to handle large datasets, provide feature importance, and deliver high accuracy makes it a valuable tool in the machine learning toolkit.

```python
#Fitting scaled X_train and y_train on Random Forest Classifier
model_rf = RandomForestClassifier()
model_rf.fit(X_train_scl, y_train)
```

```python
#Accuracy of the model on training and testing data
 
print("Training Accuracy :", model_rf.score(X_train_scl, y_train))
print("Testing Accuracy :", model_rf.score(X_test_scl, y_test))
```

```python
#Predicting on the test set
y_preds = model_rf.predict(X_test_scl)
```

```python
#Confusion Matrix
cm = confusion_matrix(y_test, y_preds)
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model_rf.classes_)
cm_display.plot()
plt.show()
```

### K-fold cross-validation
a powerful and commonly used technique in machine learning to evaluate the performance of a model. It involves partitioning the data into K subsets or "folds" and performing training and testing K times, each time using a different fold as the test set and the remaining folds as the training set.

```python
accuracies = cross_val_score(estimator = model_rf, X = X_train_scl, y = y_train, cv = 10)

print("Accuracy :", accuracies.mean())
print("Standard Variance :", accuracies.std())
```

Applying grid search to get the optimal parameters on random forest

```python
params = {
    'bootstrap': [True],
    'max_depth': [80, 100],
    'min_samples_split': [8, 12],
    'n_estimators': [100, 300]
}
```

```python
cv_object = StratifiedKFold(n_splits = 2)

grid_search = GridSearchCV(estimator = model_rf, param_grid = params, cv = cv_object, verbose = 0, return_train_score = True)
grid_search.fit(X_train_scl, y_train.ravel())
```

```python
#Getting the best parameters from the grid search

print("Best Parameter Combination : {}".format(grid_search.best_params_))
```

```python
print("Cross validation mean accuracy on train set : {}".format(grid_search.cv_results_['mean_train_score'].mean()*100))
print("Cross validation mean accuracy on test set : {}".format(grid_search.cv_results_['mean_test_score'].mean()*100))
print("Accuracy score for test set :", accuracy_score(y_test, y_preds))
```
